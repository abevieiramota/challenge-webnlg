```python
import os

os.sys.path.insert(0, '../script')
from webnlg import WebNLGCorpus
import pandas as pd
```

```python
train_dev = WebNLGCorpus.load(dataset=['train', 'dev'])
```

```python
size_1 = train_dev.subset(ntriples=[1])
size_lt1 = train_dev.subset(ntriples=[2, 3, 4, 5, 6, 7])
```

```python
lt1_plus_entry = pd.merge(size_lt1.mdf, size_lt1.edf)
d = pd.merge(size_1.mdf, lt1_plus_entry, on=['mtext'], suffixes=('_1', '_lt1'))
```

```python
d.sample().T
```

```python
counts = d.groupby(['mtext', 'ntriples']).size().unstack().fillna(0)

counts.head()
```
